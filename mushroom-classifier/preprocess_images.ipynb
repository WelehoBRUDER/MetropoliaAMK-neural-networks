```python
# This script was generated by Gemini to split the dataset before augmentation
import splitfolders
import os

# Paths - Change these to match your setup
input_folder = '../datasets/mushroom'      # Your original, un-augmented data
output_folder = '../datasets/mushroom_split' # Where the split data will go

# Split with a ratio.
# train=0.7, validation=0.15, test=0.15
splitfolders.ratio(input_folder, output=output_folder,
                   seed=1337, ratio=(.7, .15, .15), 
                   group_prefix=None, move=False) # move=False copies the files

print("Split complete. New directory structure created at:", output_folder)
```

```python
# This script was generated by Gemini when diagnosing the kernel crashes
import tensorflow as tf
from keras.utils import img_to_array, load_img, save_img
from keras.preprocessing.image import ImageDataGenerator

# Configuration
input_dir = '../datasets/mushroom_split/train'
output_dir = '../datasets/mushroom_augmented'

# This object handles the math for the transformations
datagen = ImageDataGenerator(
    rotation_range=15,    # Reduce rotation slightly to minimize gaps
    zoom_range=[0.8, 1.0], # Force a slight zoom-in
    horizontal_flip=True,
    fill_mode='constant', # Fill with a solid color
    cval=0                # Black background
)

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

print("Starting augmentation... this may take a few minutes.")

for category in os.listdir(input_dir):
    category_path = os.path.join(input_dir, category)
    save_path = os.path.join(output_dir, category)
    os.makedirs(save_path, exist_ok=True)
    
    for img_name in os.listdir(category_path):
        # Load and convert to array
        img = load_img(os.path.join(category_path, img_name))
        x = img_to_array(img)
        x = x.reshape((1,) + x.shape) # Add batch dimension
        
        # Save original and 2 augmented versions
        i = 0
        for batch in datagen.flow(x, batch_size=1, save_to_dir=save_path, 
                                  save_prefix='aug', save_format='jpeg'):
            i += 1
            if i > 2: break 

print(f"Done! Your augmented dataset is at: {output_dir}")
```
