# Assignment 2
This fully connected neural network predicts the median house value in California districts.  
The goal is to experiment with different regularization methods and to try out different model architectures.

```python
import matplotlib.pyplot as plt
import numpy as np
from keras.models import Sequential
from keras.layers import Input, Dense
from keras.datasets import california_housing

(inputs, targets), (x_test, y_test) = california_housing.load_data(version='large', 
                                        test_split=0.2)
print(inputs.shape, targets.shape)
print(x_test.shape, y_test.shape)
```

## Data preprocessing
The dataset is split into training, validation and testing datasets.  
Values are then standardized for the model.

```python
import math
# Get 80% of all inputs, since targets is the same size this also applies to it
split = math.floor(inputs.shape[0] * 0.8)
x_train = inputs[:split]
y_train = targets[:split]
x_val = inputs[split:]
y_val = targets[split:]
```

```python
mean = x_train.mean(axis=0) # mean and standard deviation computed from training set only
std = x_train.std(axis=0)
x_train -= mean
x_train /= std
x_test -= mean
x_test /= std
x_val -= mean
x_val /= std
y_train /= 1e5
y_test /= 1e5
y_val /= 1e5
```

## Testing model architecture
The goal is to try out a few different models and to see how they perform:  
128x64, 64x64 and 64x32. With the size of the inputs, a larger model doesn't make much sense.  
A few different epoch numbers were tried (40-200), but the most consistent results were achieved with 100.

```python
# define variables for reuse in other models
n_epochs = 100
n_batch = 32
# starting with a fairly large model
neurons1_n = 128
neurons2_n = 64

# define the model architecture
model = Sequential([
    Input(shape=(8,)),
    Dense(neurons1_n, activation='relu'),
    Dense(neurons2_n, activation='relu'),
    Dense(1)
    ])

# configure the learning algorithm
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
history = model.fit(x_train, y_train, epochs=n_epochs, batch_size=n_batch, validation_data=(x_val, y_val))
```

```python
plt.plot(history.history['mae'], label='train MAE')
plt.plot(history.history['val_mae'], label='val MAE')
plt.legend() 
plt.title("128x64 layers")
plt.xlabel('Epoch')
plt.ylabel('Mean absolute error (in 100000 $)')
```

```python
# Try slightly smaller
neurons1_n = 64
neurons2_n = 64

# define the model architecture
model = Sequential([
    Input(shape=(8,)),
    Dense(neurons1_n, activation='relu'),
    Dense(neurons2_n, activation='relu'),
    Dense(1)
    ])

# configure the learning algorithm
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
history = model.fit(x_train, y_train, epochs=n_epochs, batch_size=n_batch, validation_data=(x_val, y_val))
```

```python
plt.plot(history.history['mae'], label='train MAE')
plt.plot(history.history['val_mae'], label='val MAE')
plt.legend() 
plt.title("64x64 layers")
plt.xlabel('Epoch')
plt.ylabel('Mean absolute error (in 100000 $)')
```

```python
# And even smaller
neurons1_n = 64
neurons2_n = 32

# define the model architecture
model = Sequential([
    Input(shape=(8,)),
    Dense(neurons1_n, activation='relu'),
    Dense(neurons2_n, activation='relu'),
    Dense(1)
    ])

# configure the learning algorithm
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
history = model.fit(x_train, y_train, epochs=n_epochs, batch_size=n_batch, validation_data=(x_val, y_val))
```

```python
plt.plot(history.history['mae'], label='train MAE')
plt.plot(history.history['val_mae'], label='val MAE')
plt.legend() 
plt.title("64x32 layers")
plt.xlabel('Epoch')
plt.ylabel('Mean absolute error (in 100000 $)')
```

The 64x64 model still shows clear overfitting, but it produces a slightly more balanced result.  
Next, a few different regularization methods will be used to handle the overfitting.

```python
# revert back to 64x64 model
neurons1_n = 64
neurons2_n = 64
```

## Early stopping
Early stopping works by monitoring changes and ending the training if values begin to stagnate.

```python
# First trying early stopping
# define the model architecture
model = Sequential([
    Input(shape=(8,)),
    Dense(neurons1_n, activation='relu'),
    Dense(neurons2_n, activation='relu'),
    Dense(1)
    ])

# configure the learning algorithm
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
from keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit(x_train, y_train, epochs=n_epochs, batch_size=n_batch, validation_data=(x_val, y_val), callbacks=[early_stopping])
```

```python
plt.plot(history.history['mae'], label='train MAE')
plt.plot(history.history['val_mae'], label='val MAE')
plt.legend()
plt.title("Early stopping (patience=10)")
plt.xlabel('Epoch')
plt.ylabel('Mean absolute error (in 100000 $)')
```

The results still show some overfitting, but this is a good start.

## L2 Regularizer
L2 regularization prevents overfitting by adding a penalty proportional to the sum of squared weights to the loss function.

```python
# Regularization
from keras.regularizers import l2

model = Sequential([
    Input(shape=(8,)),
    Dense(neurons1_n, activation='relu', kernel_regularizer=l2(0.001)),
    Dense(neurons2_n, activation='relu', kernel_regularizer=l2(0.001)),
    Dense(1)
])
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
history = model.fit(x_train, y_train, epochs=n_epochs, batch_size=n_batch, validation_data=(x_val, y_val))
```

```python
plt.plot(history.history['mae'], label='train MAE')
plt.plot(history.history['val_mae'], label='val MAE')
plt.legend()
plt.title("L2 Regularizer (0.001)")
plt.xlabel('Epoch')
plt.ylabel('Mean absolute error (in 100000 $)')
```

It seems that the L2 regularization did not much improve the overfitting issue.

## Dropout
Dropout is a simple method that prevents overfitting by omitting (set to 0) a certain percentage of the neuron connections per layer.

```python
from keras.layers import Dropout

model = Sequential([
    Input(shape=(8,)),
    Dense(neurons1_n, activation='relu'),
    Dropout(0.2),
    Dense(neurons2_n, activation='relu'),
    Dropout(0.1),
    Dense(1)
    ])

# configure the learning algorithm
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
history = model.fit(x_train, y_train, epochs=n_epochs, batch_size=n_batch, validation_data=(x_val, y_val))
```

```python
plt.plot(history.history['mae'], label='train MAE')
plt.plot(history.history['val_mae'], label='val MAE')
plt.legend()
plt.title("Dropout")
plt.xlabel('Epoch')
plt.ylabel('Mean absolute error (in 100000 $)')
```

Dropout seems to have a significant effect on the model, as overfitting is largely gone already.

## Batch normalization
Batch normalization adjusts inputs to each layer by re-centering them around zero and re-scaling them.

```python
from keras.layers import BatchNormalization
model = Sequential([
    Input(shape=(8,)),
    Dense(neurons1_n, activation='relu'),
    BatchNormalization(),
    Dense(neurons2_n, activation='relu'),
    BatchNormalization(),
    Dense(1)
    ])

# configure the learning algorithm
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
history = model.fit(x_train, y_train, epochs=n_epochs, batch_size=n_batch, validation_data=(x_val, y_val))
```

```python
plt.plot(history.history['mae'], label='train MAE')
plt.plot(history.history['val_mae'], label='val MAE')
plt.legend()
plt.title("Batch normalization")
plt.xlabel('Epoch')
plt.ylabel('Mean absolute error (in 100000 $)')
```

The result is similar to dropout, suggesting that they are the most effective.

The most promising methods seem to be dropout and batch normalization.  
The final model will use both methods, and also implement early stopping to find the optimal epochs.

## Finalized model
The final model uses 64x64 layers with Dropout(0.2) in the first layer and BatchNormalization() on both.  
The expected result is a model with no / minimal signs of overfitting.

```python
model = Sequential([
    Input(shape=(8,)),
    Dense(neurons1_n, activation='relu'),
    Dropout(0.2),
    BatchNormalization(),
    Dense(neurons2_n, activation='relu'),
    BatchNormalization(),
    Dense(1)
    ])

# configure the learning algorithm
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
```

```python
x_train_full = np.concatenate((x_train, x_val), axis=0) # join train + val sets
y_train_full = np.concatenate((y_train, y_val), axis=0)

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
history = model.fit(x_train_full, y_train_full, epochs=n_epochs, batch_size=n_batch, validation_data=(x_val, y_val), callbacks=[early_stopping])
```

```python
plt.plot(history.history['mae'], label='train MAE')
plt.plot(history.history['val_mae'], label='val MAE')
plt.legend()
plt.title("Finalized model")
plt.xlabel('Epoch')
plt.ylabel('Mean absolute error (in 100000 $)')
```

```python
loss, acc = model.evaluate(x_test, y_test, verbose=0)
print(f"Loss: {loss}\nAccuracy: {acc}")
```

Overall, the regularization was able to address overfitting quite well.  
The final accuracy of the model came out to be only 38% which isn't great.  
Since the goal of this task was not necessarily an accurate model but combatting overfitting, the final result is still acceptable.
